```python
import re
import math
```

```python
with open('/home/user/Desktop/master_i_margarita.txt','rb') as f:
    binary = f.read()
```

```python
text = binary.decode('utf-8')

#replace some letters according to lab recommendations
text = text.replace('ъ','ь')
text = text.replace('ё','е')
```

```python
#only letters
text_filtered = re.sub(r'[^а-яА-Я]', '', text).lower()
#letters + whitespace
text_filtered_wspace = re.sub(r'[^\ а-яА-Я]', '', text).lower()
```

```python
#russian alphabet without 'ъ' and 'ё' + whitespace; 32 chars
russian_alphabet_wspace = 'абвгдежзийклмнопрстуфхцчшщыьэюя '
```

```python
monogramm_dict = dict()
monogramm_dict_wspace = dict()
```

```python
#initialize two dictionaries with pairs 'alphabet char':'number of encounters'
for char in russian_alphabet_wspace:
    monogramm_dict.update({ char : 0 })
    
for char in russian_alphabet_wspace:
    monogramm_dict_wspace.update({ char : 0 })
```

```python
#loop through text & count encounters
for char in text_filtered:
        monogramm_dict[char] += 1
        
for char in text_filtered_wspace:
        monogramm_dict_wspace[char] += 1
```

```python
#get length of text in total
length = len(text_filtered)
length_wspace = len(text_filtered_wspace)
```

```python
h_mono = 0
h_mono_wspace = 0

r_mono = 0
r_mono_wspace = 0
```

```python
for key, val in monogramm_dict.items():
    #remove chars with zero encounters 
    if val != 0:
        #calculate frequency of char (encounter_number/total_length)
        mono_frequency = val / length
        #calculate H according to the formula
        h_mono -= mono_frequency * math.log2(mono_frequency)
        print('{} | {:6d} | {:0.5f}'.format(key, val, mono_frequency)) 

        
r_mono = 1 - (h_mono) / math.log2(len(russian_alphabet_wspace))
print('Entropy: {:0.5f}; Redundancy {:0.5f}'.format(h_mono, r_mono))

for key, val in monogramm_dict_wspace.items():
    if val != 0:
        mono_frequency_wspace = val / length_wspace
        h_mono_wspace -= mono_frequency_wspace * math.log2(mono_frequency_wspace)
        print('{} | {:6d} | {:0.5f}'.format(key, val, mono_frequency_wspace))  

        
r_mono_wspace = 1 - (h_mono_wspace) / math.log2(len(russian_alphabet_wspace))   
print('Entropy: {:0.5f}; Redundancy {:0.5f}'.format(h_mono_wspace, r_mono_wspace))
```

```python
bigramm_dict_1 = dict()
bigramm_dict_2 = dict()

bigramm_dict_wspace_1 = dict()
bigramm_dict_wspace_2 = dict()
```

```python
#initialize two dicts for bigramms with pairs 'bigramm':'number of encounters'
for char_1 in russian_alphabet_wspace:
    for char_2 in russian_alphabet_wspace:
        key = char_1 + char_2
        bigramm_dict_1.update({ key : 0 })
        bigramm_dict_2.update({ key : 0 })
        
for char_1 in russian_alphabet_wspace:
    for char_2 in russian_alphabet_wspace:
        key = char_1 + char_2
        bigramm_dict_wspace_1.update({ key : 0 })
        bigramm_dict_wspace_2.update({ key : 0 })
```

```python
#initialize variables to store total count of bigramms in text
bigramm_count_1 = 0
bigramm_count_2 = 0

bigramm_count_wspace_1 = 0
bigramm_count_wspace_2 = 0
```

```python
#initialize temp variables to store text & boolean for loop condition
temp = text_filtered
text_present = True

temp_wspace = text_filtered_wspace
text_present_wspace = True
```

```python
#text step = 1 char
#while text is present
while text_present:
    #get first two chars (bigramm)
    bigramm = temp[:2]
    #move 1 char forward
    temp = temp[1:]
    
    if len(bigramm) < 2:
        #if we read 1 char or 0 = text has ended
        text_present = False
    else:
        #bigramms count ++
        bigramm_count_1 += 1
        #increment the encounter number for the bigramm
        bigramm_dict_1[bigramm] += 1
        
while text_present_wspace:
    bigramm = temp_wspace[:2]
    temp_wspace = temp_wspace[1:]
    
    if len(bigramm) < 2:
        text_present_wspace = False
    else:
        bigramm_count_wspace_1 += 1
        bigramm_dict_wspace_1[bigramm] += 1
```

```python
#reinitialize variables
temp = text_filtered
text_present = True

temp_wspace = text_filtered_wspace
text_present_wspace = True
```

```python
#text step = 2 chars
#while text is present
while text_present:
    #get first two chars of text (bigramm)
    bigramm = temp[:2]
    #move 2 chars forward
    temp = temp[2:]
    
    if len(bigramm) < 2:
        #if we read 1 char or 0 = text has ended
        text_present = False
    else: 
        #bigramms count++
        bigramm_count_2 += 1
        #increment the encounter number for the bigramm
        bigramm_dict_2[bigramm] += 1
        
while text_present_wspace:
    bigramm = temp_wspace[:2]
    temp_wspace = temp_wspace[2:]
    
    if len(bigramm) < 2:
        text_present_wspace = False
    else:
        bigramm_count_wspace_2 += 1
        bigramm_dict_wspace_2[bigramm] += 1
```

```python
h_bi_1 = 0
h_bi_2 = 0

h_bi_wspace_1 = 0
h_bi_wspace_2 = 0
```

```python
r_bi_1 = 0
r_bi_2 = 0

r_bi_wspace_1 = 0
r_bi_wspace_2 = 0
```

```python
for key, val in bigramm_dict_1.items():
    if val != 0:  
        bi_1_frequency = val / bigramm_count_1 
        h_bi_1 -= bi_1_frequency * math.log2(bi_1_frequency) / 2
        print('{} | {:6d} | {:0.5f}'.format(key, val, bi_1_frequency))
        
r_bi_1 = 1 - (h_bi_1) / math.log2(len(russian_alphabet_wspace))
print('Entropy: {:0.5f}; Redundancy {:0.5f}'.format(h_bi_1, r_bi_1))


for key, val in bigramm_dict_wspace_1.items():
    if val != 0:
        bi_1_wspace_frequency = val/bigramm_count_wspace_1 
        h_bi_wspace_1 -= bi_1_wspace_frequency * math.log2(bi_1_wspace_frequency) / 2
        print('{} | {:6d} | {:0.5f}'.format(key, val, bi_1_wspace_frequency))

r_bi_wspace_1 = 1 - (h_bi_wspace_1)/math.log2(len(russian_alphabet_wspace))
print('Entropy: {:0.5f}; Redundancy {:0.5f}'.format(h_bi_wspace_1, r_bi_wspace_1))
```

```python
for key, val in bigramm_dict_2.items():
    if val != 0:
        bi_2_frequency = val / bigramm_count_2
        h_bi_2 -= bi_2_frequency * math.log2(bi_2_frequency) / 2 
        print('{} | {:6d} | {:0.5f}'.format(key,val,bi_2_frequency))
    
r_bi_2 = 1 - (h_bi_2)/math.log2(len(russian_alphabet_wspace))
print('Entropy: {:0.5f}; Redundancy {:0.5f}'.format(h_bi_2, r_bi_2))


for key, val in bigramm_dict_wspace_2.items():
    if val != 0:
        bi_2_wspace_frequency = val/bigramm_count_wspace_2 
        h_bi_wspace_2 -= bi_2_wspace_frequency * math.log2(bi_2_wspace_frequency) / 2
        print('{} | {:6d} | {:0.5f}'.format(key, val, bi_2_wspace_frequency))
       
r_bi_wspace_2 = 1 - (h_bi_wspace_2)/math.log2(len(russian_alphabet_wspace))
print('Entropy: {:0.5f}; Redundancy {:0.5f}'.format(h_bi_wspace_2, r_bi_wspace_2))
```
